```python
#Defining all the needed values

#Those will be explicidly computed
z,x,y,mu = var('z,x,y,mu')

#Those will have an imput as a basis
avg_m = 1
delta_m = 0
z1 = 0.25 #sage doesn't like fractions...
```

```python
#Sets of equations as seen in the theory

#Zetas
zeta = x + I*y
zeta_bar = x - I*y

#I here denotes a square root of -1, bar stands for conjugates
```

```python
#Looking for expression n.1 which will be used to denote polynomial n.1

z_bar(z) = zeta_bar - (avg_m + delta_m)/(z1 - z) - (avg_m - delta_m)/(-z1-z)
L(z) = zeta - z - (avg_m + delta_m)/(z1 - z_bar(z)) - (avg_m - delta_m)/(-z1 - z_bar(z))

print(L(z))
```

```python
exp1 = L(z).numerator(normalize=true)
print(exp1)
```

```python
#Looking for expression n.2 which will be used to denote polynomial n.1

#derivative of z
derivative_z(z) = (delta_m + avg_m)/(z1 - z_bar(z))^2 + (avg_m - delta_m)/(z1+z_bar(z))^2

#and its conjugate
derivative_z_bar(z) = (delta_m + avg_m)/(z1 - z)^2 + (avg_m - delta_m)/(z1+z)^2

#magnification polynomial
magnification(z) = mu*(1-derivative_z(z)*derivative_z_bar(z))-1
print(magnification(z))
```

```python
exp2 = magnification(z).numerator(normalize=true)
print(exp2)
```

```python
#In SageMath we have to transform our symbolic expressions into polynomials via polynomial rings.
#Therefore, we define 2: one which holds constants of a polynomial(still symbolic) and one which holds z:

R = PolynomialRing(CC,'x,y,mu')
print(R)

#CDF denoted complex double field. It is less precise than standard complex field with 53 digits of precision, however, faster.

R.inject_variables()

S.<z> = R[]
print(S)

#We create out polynomials by recreating out expressions through rings.
poly1 = exp1.polynomial(None,ring=S)
poly2 = exp2.polynomial(None,ring=S)

#Now the polynomials are based on rings
poly1.parent()
```

```python
#We can compute the resultant. It takes around 10s seconds. For non-special case, 30s

resultant = poly1.resultant(poly2)
```

```python
#Solving the resultant for mu

t1(mu) = resultant.subs({x:-0.25,y:0.-25})

roots = solve([t1(mu)==0],mu,to_poly_solve="force")
stop = len(roots)
log(sum([abs(roots[i].rhs().n()) for i in range(stop)]))
```

```python
#here x values = a, y values = b to avoid confusion with previously defined ones

a,b = var('a,b')

def g(a,b):
    temp(mu) = resultant.subs({x:a,y:b})
    roots = solve([temp(mu)==0],mu,to_poly_solve="force")
    mag = sum([abs(roots[i].rhs().n()) for i in range(len(roots))])
    return log(mag)
```
